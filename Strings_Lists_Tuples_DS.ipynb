# List In data structures
1. list is collection of elements enclosed by []
2. it's allowed duplicate elements
3. each element seperated by ,
4. list is mutable - editable 

```python
# create list
names=["shiva","vikas","divya","keshav"]
ages=[25,26,24,27,27,25]
print(ages)
print(names)
print(type(names))
print(type(ages))
```

```python
#concatenation of list
a1=[10,20,30,40,50]
a2=[10.25,12.56,13.58]
a3=a1+a2
print(a3)
```

```python
#repeate list
info=["shiva","divya",10,20,30,1000.25,12500.26]
print(info)
print(info*5)
```

```python
#accessing elements from list
#index based
info=["shiva","divya",10,20,30,1000.25,12500.26]
print(info)
print(info[1])
print(info[-6])
```

```python
info=["shiva","divya",10,20,30,1000.25,12500.26]
print(info)
print(info[2:5])
print(info[0:2])
print(info[5:7])
```

```python
#methods or functions support by List
print(dir(list))
```

```python
#methods with description
print(help(list))
```

```python
#append()
names=["shiva","vikram","divya"]
names.append("chandu")
print(names)
```

```python
#extend()
names=["shiva","vikra","divya"]
names.extend(["chandu","vikas"])
print(names)
```

```python
#insert()
names=["shiva","vikram","divya"]
names.insert(1,"vikas")
print(names)
```

```python
#pop()
numbers=[10,20,30,40,50]
print(numbers)
numbers.pop(2)
print(numbers)
```

```python
#remove()
ages=[25,26,27,28,29,23]
print(ages)
ages.remove(28)
print(ages)
```

```python
#sorting techniques sort() & reverse
numbers=[5,1,7,9,4,6,3,2]
print(numbers)
numbers.sort(reverse=False)
print(numbers)
```
