# dictionaries in Data Structures
1. dict is collection of items enclosed by {}
2. here item is a key:value pair
3. keys are unique
4. each item seperateb by ,
5. dict is mutable - editable

```python
#create dictionaries
info={"name":"shiva","age":25,"address":"hyderabad"}
print(info)
print(type(info))
```

```python
# methods or functions
print(dir(dict))
```

```python
#methods with description
print(help(dict))
```

```python
#keys(),values(),items()
info={"name":"shiva","age":36,"address":"hyderabad"}
print(info)
print(info.keys())
print(info.values())
print(info.items())
```

```python
# update - insert or update items to dict
info={"name":"shiva","age":26,"address":"hyd"}
print(info)
info.update({"address":"banglore"})
print(info)
```

```python
#pop method
info={"name":"shiva","age":26,"address":"hyderabad"}
print(info)
info.pop("age")
print(info)
```
